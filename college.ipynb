```python
import pandas as pd

college = pd.read_csv('College.csv', index_col=0)
college.head()
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']])
```

```python
college.boxplot(column='Outstate', by='Private')
```

```python
college['Elite'] = pd.cut((college['Top10perc'] / college['Enroll']), [0, 0.5, 1], labels=['No', 'Yes'])
college['Elite'].value_counts()
```

```python
college.boxplot(column='Outstate', by='Elite')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2)
college.plot.hist(y='Outstate', bins=20, ax=ax[0, 0])
college.plot.hist(y='Top10perc', bins=10, ax=ax[0, 1])
college.plot.hist(y='Enroll', bins=30, ax=ax[1, 0])
college.plot.hist(y='Apps', bins=30, ax=ax[1, 1])
```
